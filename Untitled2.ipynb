```python
import numpy as np
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LeakyReLU, Reshape, Flatten
from tensorflow.keras.optimizers import Adam

# Load and normalize MNIST data
(X_train, _), _ = mnist.load_data()
X_train = (X_train - 127.5) / 127.5  # Scale to [-1, 1]
X_train = X_train.reshape(-1, 784)

# Optimizer
opt = Adam(0.0002, 0.5)

# --- Generator function ---
def build_generator():
    model = Sequential([
        Dense(128, input_dim=100),
        LeakyReLU(0.2),
        Dense(784, activation='tanh'),  # Output shape: 28x28 = 784
    ])
    return model

# --- Discriminator function ---
def build_discriminator():
    model = Sequential([
        Dense(128, input_shape=(784,)),
        LeakyReLU(0.2),
        Dense(1, activation='sigmoid')  # Binary output: real (1) or fake (0)
    ])
    model.compile(optimizer=opt, loss='binary_crossentropy', metrics=['accuracy'])
    return model

# Build models
generator = build_generator()
discriminator = build_discriminator()

# --- Train GAN ---
noise = np.random.normal(0, 1, (128, 100))          # Random noise
fake_images = generator.predict(noise)              # Generate fake images

real_images = X_train[np.random.randint(0, X_train.shape[0], 128)]  # Sample real images

# Labels for training
real_labels = np.ones((128, 1))
fake_labels = np.zeros((128, 1))

# Train discriminator
d_loss_real = discriminator.train_on_batch(real_images, real_labels)
d_loss_fake = discriminator.train_on_batch(fake_images, fake_labels)

# Print discriminator performance
print("Discriminator accuracy on real:", d_loss_real[1]*100)
print("Discriminator accuracy on fake:", d_loss_fake[1]*100)
```

```python
import torch
import torch.nn as nn
import math

# ----------------------------
# Multi-Head Attention
# ----------------------------
class MultiHeadAttention(nn.Module):
    def __init__(self, d_model, num_heads):
        super().__init__()
        self.num_heads = num_heads
        self.head_dim = d_model // num_heads

        assert self.head_dim * num_heads == d_model

        self.qkv = nn.Linear(d_model, d_model * 3)
        self.fc_out = nn.Linear(d_model, d_model)

    def forward(self, x):
        B, T, C = x.shape

        qkv = self.qkv(x)  # (B, T, 3C)
        qkv = qkv.reshape(B, T, 3, self.num_heads, self.head_dim)
        qkv = qkv.permute(2, 0, 3, 1, 4)

        Q, K, V = qkv[0], qkv[1], qkv[2]

        scores = (Q @ K.transpose(-2, -1)) / math.sqrt(self.head_dim)
        attn = torch.softmax(scores, dim=-1)

        out = attn @ V
        out = out.transpose(1, 2).reshape(B, T, C)

        return self.fc_out(out)

# ----------------------------
# Feed Forward
# ----------------------------
class FeedForward(nn.Module):
    def __init__(self, d_model, hidden_dim):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(d_model, hidden_dim),
            nn.ReLU(),
            nn.Linear(hidden_dim, d_model)
        )

    def forward(self, x):
        return self.net(x)

# ----------------------------
# Transformer Block
# ----------------------------
class TransformerBlock(nn.Module):
    def __init__(self, d_model, num_heads, hidden_dim):
        super().__init__()
        self.attn = MultiHeadAttention(d_model, num_heads)
        self.ff = FeedForward(d_model, hidden_dim)

        self.norm1 = nn.LayerNorm(d_model)
        self.norm2 = nn.LayerNorm(d_model)

    def forward(self, x):
        x = x + self.attn(self.norm1(x))
        x = x + self.ff(self.norm2(x))
        return x

# ----------------------------
# Full Transformer Encoder
# ----------------------------
class TransformerEncoder(nn.Module):
    def __init__(self, vocab_size, d_model, num_heads, num_layers, hidden_dim, max_len=100):
        super().__init__()

        self.embedding = nn.Embedding(vocab_size, d_model)
        self.pos_embedding = nn.Parameter(torch.randn(1, max_len, d_model))

        self.layers = nn.ModuleList([
            TransformerBlock(d_model, num_heads, hidden_dim)
            for _ in range(num_layers)
        ])

        self.fc_out = nn.Linear(d_model, vocab_size)

    def forward(self, x):
        B, T = x.shape

        x = self.embedding(x) + self.pos_embedding[:, :T, :]

        for layer in self.layers:
            x = layer(x)

        return self.fc_out(x)

# ----------------------------
# Test Run
# ----------------------------
model = TransformerEncoder(
    vocab_size=1000,
    d_model=128,
    num_heads=8,
    num_layers=2,
    hidden_dim=256
)

x = torch.randint(0, 1000, (32, 20))  # (batch, seq_len)
out = model(x)

print(out.shape)  # (32, 20, vocab_size)
```

Observations – Transformer Model
The Transformer encoder successfully processes input sequences using self-attention, capturing long-range dependencies better than RNNs.
Multi-head attention allows the model to focus on different parts of the sequence simultaneously, improving contextual understanding.
Positional embeddings are essential—without them, the model cannot understand word order.
The output tensor shape (batch_size, seq_len, vocab_size) confirms correct sequence-to-sequence mapping.
Training convergence depends heavily on:
learning rate
number of layers
attention heads
Compared to traditional models, Transformer shows:
faster parallel computation
better scalability
🔹 Observations – BERT Model Tasks

(Using BERT)

🔸 1. Next Word / Mask Prediction
The model predicts contextually relevant words for [MASK].
Predictions vary based on sentence structure and surrounding words.
Output includes multiple candidates with probability scores.
Works best when input text is lowercase (for uncased model).
🔸 2. Missing Word Filling
BERT accurately fills missing words using bidirectional context.
It performs better than unidirectional models since it considers both left and right context.
Accuracy depends on how meaningful the sentence context is.
🔸 3. Sentiment Classification
The model correctly classifies text into positive/negative sentiment.
High confidence scores (>0.9) indicate strong prediction reliability.
Performance is strong even without additional training (pretrained model).
May struggle with sarcasm or ambiguous language.
🔹 Observations – Vision Transformer (ViT)
The Vision Transformer processes images by splitting them into fixed-size patches.
Each patch is treated like a token, similar to words in NLP.
The class token ([CLS]) aggregates information for final classification.
Model output shape (batch_size, num_classes) confirms correct classification pipeline.
ViT requires large datasets for best performance compared to CNNs.
Positional embeddings are crucial to maintain spatial information of patches.
Compared to CNN:
better global feature understanding
less inductive bias
higher data requirement
🔹 Overall Observation
Transformer-based models unify NLP and Vision tasks under a single architecture.
Self-attention is the core mechanism enabling:
contextual understanding (BERT)
sequence modeling (Transformer)
image representation (ViT)
Pretrained models significantly reduce training time and improve performance.
These models are highly scalable but computationally expensive.
